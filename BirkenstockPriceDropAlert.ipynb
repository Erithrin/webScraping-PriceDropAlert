```python
#we now write the actual code for the deal price alert to work

def checkPrice():
    url = 'https://www.birkenstock.com/us/gizeh-birko-flor/gizeh-electricmetallic-birkoflor-0-eva-w_3651.html#sz=48'
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
    page = req.get(url, headers = headers)

    soup = BeautifulSoup(page.content, "html.parser")
    soup2 = BeautifulSoup(soup.prettify(),"html.parser")
    price = soup2.find("span", class_="price-standard").get_text()
    print(price)
    
    model = soup2.find("span", class_="product-modelname").get_text()
    print(model)
    
    product_name = soup2.find("span",class_ = "product-shortname").get_text()
    print(product_name)

    price = price.strip()[1:]
    price = int(float(price))
    model = model.strip()
    product_name = product_name.strip()
    
    #creating a timestamp for when the data was collected
    import datetime
    import csv
    from pathlib import Path
    
    today = datetime.date.today()
    print(today)
    
    header = ["model","product_detail","price","date"]
    data = [model, product_name, price, today]
    file_path = Path("Birkenstock-deal-alert.csv")
    if not file_path.exists():
        with open('Birkenstock-deal-alert.csv', 'w' ,newline='', encoding='UTF8') as f:
            writer = csv.writer(f)
            writer.writerow(header)
            writer.writerow(data)
    else:
        with open('Birkenstock-deal-alert.csv', 'a+', newline='',encoding='UTF8') as f:
            writer = csv.writer(f)
            writer.writerow(data)
            
    df  = pd.read_csv(file_path)
    print(df)

    if price<=99:
        send_mail()
   
```

```python
import time
while(True):
    checkPrice()
    time.sleep(60)


```
